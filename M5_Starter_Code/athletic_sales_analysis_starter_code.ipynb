### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
df_combined
```

```python
#30 Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_sales = df_combined[df_combined['product'] == "Women's Athletic Footwear"]
womens_athletic_footwear_sales
```

#### Using `groupby`

```python
#31 Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
#32 Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_units_sold = womens_athletic_footwear_sales.groupby(
    ["retailer", "region", "state", "city"]
)["units_sold"].sum().reset_index()

womens_footwear_units_sold.rename(
    columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
#33 Show the top 5 results.
top_5_womens_footwear_units_sold = womens_footwear_units_sold.sort_values(
    'Womens_Footwear_Units_Sold', ascending=False).head(5)
top_5_womens_footwear_units_sold



```

#### Using `pivot_table`

```python
#34 Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_pivot = pd.pivot_table(womens_athletic_footwear_sales, 
                                       index=["retailer", "region", "state", "city"], 
                                       values="units_sold", 
                                       aggfunc="sum")


#35 Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_pivot = womens_footwear_pivot.reset_index()
womens_footwear_pivot = womens_footwear_pivot.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
#36 Show the top 5 results.
top_5_womens_footwear = womens_footwear_pivot.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
top_5_womens_footwear



```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
#37 Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
daily_sales = pd.pivot_table(
    womens_athletic_footwear_sales,
    index='invoice_date', 
    values='total_sales', 
    aggfunc='sum'
)

#38 Optional: Rename the "total_sales" column to "Total Sales"
daily_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

#39 Show the table.
daily_sales
```

```python
#40 Resample the pivot table into daily bins, and get the total sales for each day.
df_combined.set_index('invoice_date', inplace=True)
daily_total_sales = df_combined['total_sales'].resample('D').sum()

#41 Sort the resampled pivot table in ascending order on "Total Sales".
sorted_daily_sales = daily_total_sales.sort_values(ascending=True)

sorted_daily_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
#42 Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = df_combined['total_sales'].resample('W').sum()
#43 Sort the resampled pivot table in ascending order on "Total Sales".
sorted_weekly_sales = weekly_sales.sort_values(ascending=False)
sorted_weekly_sales_df = sorted_weekly_sales.reset_index()
sorted_weekly_sales_df.columns = ['Week_Start_Date', 'Total_Sales']
sorted_weekly_sales.head()
```
